```python
x=ratings_with_name.groupby('User-ID').count()['Book-Rating']>200
padhe_likhe_users=x[x].index
padhe_likhe_users
```

```python
filtered_ratings=ratings_with_name[ratings_with_name['User-ID'].isin(padhe_likhe_users)]
filtered_ratings
```

```python
y=filtered_ratings.groupby('Book-Title').count()['Book-Rating']>=50
famous_books=y[y].index
final_ratings=filtered_ratings[filtered_ratings['Book-Title'].isin(famous_books)]
```

```python
pt=final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_scores=cosine_similarity(pt)
```

```python
similarity_scores
```

```python
def recommend(book_name):
    index=np.where(pt.index==book_name)[0][0]
    # distances=similarity_scores(index)
    similar_items=sorted(list(enumerate(similarity_scores[index])), key=lambda x:x[1],reverse=True)[1:6]
    data=[]
    for i in similar_items:
        item=[]
        print(pt.index[i[0]])
        temp_df=books[books['Book-Title']==pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))

        data.append(item)
    return data
```

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import TruncatedSVD
from sklearn.metrics import mean_squared_error
from scipy.sparse import coo_matrix

# ratings_with_name DataFrame with ['User-ID','Book-Title','Book-Rating']

# Map users and items to numeric IDs
user_codes = ratings_with_name['User-ID'].astype('category').cat.codes
item_codes = ratings_with_name['Book-Title'].astype('category').cat.codes
ratings = ratings_with_name['Book-Rating'].values

# Build sparse user-item matrix
sparse_matrix = coo_matrix((ratings, (user_codes, item_codes)))

# Apply SVD on sparse matrix
svd = TruncatedSVD(n_components=20, random_state=42)
matrix_reduced = svd.fit_transform(sparse_matrix)

# Reconstruct predicted ratings only for non-zero entries
predicted_ratings = np.sum(matrix_reduced[sparse_matrix.row] * svd.components_[:, sparse_matrix.col].T, axis=1)

# Compute RMSE on actual ratings
rmse = np.sqrt(mean_squared_error(ratings, predicted_ratings))
print("RMSE:", rmse)
```

```python
# Mapping indices back to original IDs / titles
user_map = dict(enumerate(ratings_with_name['User-ID'].astype('category').cat.categories))
item_map = dict(enumerate(ratings_with_name['Book-Title'].astype('category').cat.categories))
```

```python
# ----------------- Hybrid Recommender -----------------

# 1️⃣ CF helper: map SVD predictions to book titles
def get_cf_scores(user_id):
    user_idx = list(user_map.keys())[list(user_map.values()).index(user_id)]
    cf_scores_array = matrix_reduced[user_idx] @ svd.components_
    cf_dict = {item_map[i]: cf_scores_array[i] for i in range(len(item_map))}
    return cf_dict

# 2️⃣ CBF helper: map similarity scores to book titles
def get_cbf_scores(book_name):
    index = np.where(pt.index == book_name)[0][0]
    similar_items = list(enumerate(similarity_scores[index]))
    cbf_dict = {pt.index[i]: score for i, score in similar_items}
    return cbf_dict

# 3️⃣ Hybrid function
def hybrid_recommend(user_id, book_name=None, alpha=0.6, top_n=10):
    # CF predictions
    cf_dict = get_cf_scores(user_id)
    
    # CBF predictions (if a seed book is provided)
    cbf_dict = get_cbf_scores(book_name) if book_name else {}
    
    # Combine scores
    all_items = set(cf_dict.keys()).union(set(cbf_dict.keys()))
    hybrid_scores = {item: alpha * cf_dict.get(item, 0) + (1 - alpha) * cbf_dict.get(item, 0)
                     for item in all_items}
    
    # Sort by hybrid score
    recommended = sorted(hybrid_scores.items(), key=lambda x: x[1], reverse=True)[:top_n]
    
    # Fetch book details using existing books DataFrame
    data = []
    for title, score in recommended:
        temp_df = books[books['Book-Title'] == title].drop_duplicates('Book-Title')
        for _, row in temp_df.iterrows():
            data.append([row['Book-Title'], row['Book-Author'], row['Image-URL-M']])
    
    return data

# ----------------- Example Usage -----------------
book_name = '1984'  # a valid book from your pt.index
user_id = 276725    # a valid user from your CF data
top_recs = hybrid_recommend(user_id, book_name)

for rec in top_recs:
    print(rec)

```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import NearestNeighbors
import pandas as pd

# Ensure no missing titles
books = books.dropna(subset=['Book-Title']).reset_index(drop=True)

# Use titles as text
books['Text'] = books['Book-Title']

# TF-IDF vectorization
tfidf = TfidfVectorizer(stop_words='english', max_features=50000)  # limit features
tfidf_matrix = tfidf.fit_transform(books['Text'])

# Find top 5 similar books
nn = NearestNeighbors(n_neighbors=5, metric='cosine', algorithm='brute')
nn.fit(tfidf_matrix)

# Map book title to index
title_to_index = {title: idx for idx, title in enumerate(books['Book-Title'].values)}

# # Example: find top 5 similar books for a given title
# book_title = "1984"
# idx = title_to_index[book_title]
# distances, indices = nn.kneighbors(tfidf_matrix[idx])

# print(f"Top 5 books similar to '{book_title}':")
# for i, similar_idx in enumerate(indices[0]):
#     print(f"{i+1}. {books['Book-Title'][similar_idx]} (distance: {distances[0][i]:.4f})")
```

```python
# ----------------- CBF Helper (TF-IDF) -----------------
def get_cbf_scores_tfidf(seed_books):
    cbf_scores = {}
    valid_books = [b for b in seed_books if b in title_to_index]
    if not valid_books:
        return cbf_scores

    for book in valid_books:
        idx = title_to_index[book]
        sims = similarity_scores[idx]
        for i, score in enumerate(sims):
            cbf_scores[books['Book-Title'].iloc[i]] = cbf_scores.get(books['Book-Title'].iloc[i], 0) + score

    # Average if multiple seed books
    cbf_scores = {k: v / len(valid_books) for k, v in cbf_scores.items()}
    return cbf_scores

def hybrid_recommend_tfidf(user_id, seed_books=None, alpha=0.6, top_n=10):
    # 1️⃣ CF scores
    cf_scores = get_cf_scores(user_id)

    # 2️⃣ CBF scores
    cbf_scores = get_cbf_scores_tfidf(seed_books) if seed_books else {}

    # 3️⃣ Combine scores
    all_items = set(cf_scores.keys()).union(cbf_scores.keys())
    hybrid_scores = {item: alpha * cf_scores.get(item, 0) + (1 - alpha) * cbf_scores.get(item, 0)
                     for item in all_items}

    # 4️⃣ Get top N
    top_titles = sorted(hybrid_scores, key=hybrid_scores.get, reverse=True)[:top_n]

    # 5️⃣ Fetch book details
    recommended_df = books[books['Book-Title'].isin(top_titles)].drop_duplicates('Book-Title')
    result = recommended_df[['Book-Title', 'Book-Author', 'Image-URL-M']].values.tolist()
    
    return result

```

```python
user_id = 276725
seed_books = ['1984']

top_recs = hybrid_recommend_tfidf(user_id, seed_books, alpha=0.6, top_n=10)

for rec in top_recs:
    print(rec)
```
